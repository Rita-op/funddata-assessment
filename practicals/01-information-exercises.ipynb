***Adapt the code above to generate a 1000 character long string with weights based on the previous two characters.***




In the following lines, the necessary code to carry out the exercise has been obtained from the Jupyter document "01-information". It should be noted that some parts of the code have been modified in order to get what is required in the exercise. The code that has been modified is explained throughout the exercise.

```python
# Make HTTP requests for internet reqources.
import urllib.request

# The URL of a text version of Alice in Wonderland.
book_url = 'https://www.gutenberg.org/files/11/11-0.txt'

# Get the book.
book = list(urllib.request.urlopen(book_url))

# Decode the lines and strip line endings.
book = [line.decode('utf-8-sig').strip() for line in book]
```

```python
# All letters and a space, which are the characters needed in the exercise.
chars = 'abcdefghijklmnopqrstuvwxyz '

# Get the whole book in one big string and lower-case it.
sbook = ''.join(book[26:]).lower()
```

```python
# Create the weights based on the previous two characters and taking into account the occurences of each character 
# in the whole book. For that, I made use of a nested dictionary that is made of three levels.
twoghts = {c: {d: {e: sbook.count(c + d + e) for e in chars} for d in chars} for c in chars}

# Show the weights.
twoghts
```

```python
import random

# Start with the pairs of characters space and e.
pairs = ' e'

# Set the number of random characters to generate. The number has been set to 999 since we need to generate 
# 1000 character long string and we already chose the first two characters to start with and the function range will
# generate N-1 characters.
N = 999

# Do the following N-1 times.
for i in range(1, N):
    # Get the weights taking into account the two previous characters in "pairs". Considering the second last character
    # we get the second level dictionary of twoghts and considering the first previous one, we get the third level dictionary.
    wt = twoghts[pairs[-2]][pairs[-1]]
    # Turn wt into a list, ordered by chars.
    wt = [wt[c] for c in chars]
    # Randomly pick the next character using those weights.
    nextc = random.choices(chars, weights=wt, k=1)[0]
    # Append the character to twos.
    pairs = pairs + nextc
    
```

```python
# Print the final result, the 1000 character long string with weights based on the previous two characters that was required
print(pairs)
```

***Explain why the log of zero is undefined.***

The logarithms were discovered in the 16th century and since then, they have been widely used. By definition, the logarithm of a given number *x* is the exponent to which the base *b* must be raised to yield the number *x*. This kind of function is closely related to the exponential function since it is considered the inverse of the exponential function.

$$ 
f(x) = log_bx \\ f(x) = b^x 
$$   

Taking the previous relationship into account, we can perceive why the log of zero is undefined since there is no number to which the base b is raised that gives us zero as a result.  For that reason, we can conclude that the logarithm of zero is not defined.

In fact, the domain of the logarithmic functions is the set of all positive numbers. In the following lines, I will make use of python code to demonstrate the above.


```python
import math

# Try to calculate the logarithm base 2 of zero:
print(math.log2(0))
```

```python
# Try to calculate the logarithm base 2 of a negative number:
print(math.log2(-4))
```

If we run the previous code, we will obtain an error (ValueError: math domain error). This demonstrates what was said in the previous lines about the logarithmic function being undefined for zero and negative numbers. On the other hand, the code shown below works since we are calculating and plotting the logarithm of positive numbers.

```python
import numpy as np
import matplotlib.pyplot as plt

# Use the arange function to get a range of floating-point numbers to define the x variable
x = np.arange(0.1,6, 0.02)

# Define the function that we need to represent graphically
y = np.log2(x)

# Display the plot and give it a title
plt.plot(x,y)
plt.title('Representation of the logarithmic function')
plt.show()
```
